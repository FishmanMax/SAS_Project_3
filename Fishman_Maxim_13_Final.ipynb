```

import csv
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

#1.1

```
data = pd.read_csv('HW1_var_13.csv', sep=';')
data = data[data.columns[1:]]  # Удаляем номер варианта.
```

```
unique_column = []
per_unique_column = []
zeros = []
per_zeros = []
na = []
per_na = []
for i in data.columns:
  unique_column.append(data[i].nunique())
  zeros.append(data[i].isin([0]).sum())
  na.append(data[i].isna().sum())

per_unique_column = unique_column / data.nunique().sum() * 100
per_unique_column = per_unique_column.round(decimals=2)

per_zeros = zeros / data.isin([0]).sum().sum() * 100
per_zeros = per_zeros.round(decimals=2)

per_na = na / data.isna().sum().sum() * 100
per_na = per_na.round(decimals=2)
```

```
#@title Текст заголовка по умолчанию
first_task = pd.DataFrame()
first_task['VARIABLE'] = data.columns
first_task['UNIQUE'] = unique_column
first_task['PERCENT_UNIQUE'] = per_unique_column
first_task['ZEROS'] = zeros
first_task['PERCENT_ZEROS'] = per_zeros
first_task['NANS'] = na
first_task['PERCENT_NANS'] = per_na
first_task
```

Let's make a portrait of a "typical" person from this database based only on numerical attributes:
The average debt-to-income ratio is quite high - about $0.4, On average it is a fairly young person of 30-35 years old with 1 child, more often without children. This table does not include people over 62 years of age. If there is a delay in payment on the loan, it is on average no more than 0.2 months.
With all that, most of them took a new loan quite recently (not more than a year ago), although they took their first one about 5-6 years ago. A certain "dependence" on loans is traced. This is also indicated by the ratio of the number of all loans (the average number is close to 6) to the number of closed loans (the average number is less than 1).
